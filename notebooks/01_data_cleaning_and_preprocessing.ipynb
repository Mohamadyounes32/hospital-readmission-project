# Hospital Readmission Project — Data Cleaning

This notebook focuses on cleaning and preprocessing hospital readmission data prior to exploratory analysis and statistical modeling.

Key objectives:

* Handle missing and unknown values
* Improve data consistency
* Prepare variables for downstream analytics
* Ensure higher analytical reliability


```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
# Load hospital readmission dataset
df = pd.read_csv(
    "/drive/hospital-readmission-project/data/diabetic_data.csv")
```

## Initial Dataset Exploration

This section examines the overall structure of the dataset, including dimensions, variable types, and summary statistics to better understand the data before preprocessing.

```python
# Initial dataset inspection
df.head()

df.shape

df.info()

df.describe()
```

## Missing Value Analysis

Healthcare datasets frequently contain incomplete or unknown information. This section identifies variables with missing values and evaluates their potential impact on downstream analysis.

```python
# Detect unknown values represented as '?'
(df == "?").sum().sort_values(ascending=False)
```

Several variables contain significant amounts of unknown or incomplete information. This is common in real-world healthcare datasets and may affect downstream statistical analysis and predictive modeling.

## Data Cleaning and Preprocessing

This section focuses on handling missing values, improving consistency, and preparing the dataset for exploratory analysis and statistical modeling.

```python
# Replace placeholder missing values with NaN
df.replace("?", np.nan, inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Drop columns with excessive missing values
df.drop(columns=["weight"], inplace=True)
```

The weight column was removed due to an extremely high proportion of missing values, which limits its analytical reliability.

```python
# Impute missing categorical values

df["race"].fillna("Unknown", inplace=True)
df["payer_code"].fillna("Unknown", inplace=True)
df["medical_specialty"].fillna(
    "Unknown",inplace=True)
```

## Data Validation

```python
# Check for duplicate records
duplicates = df.duplicated().sum()

print("Duplicate rows:", duplicates)

df.duplicated().sum()

df.drop_duplicates(inplace=True)
```

Duplicate records were evaluated and removed to improve dataset consistency and reduce the risk of biased analytical results.

```python
# Validate final dataset structure after preprocessing
df.info()

df.isnull().sum()

df.head()
```

## Export Cleaned Dataset

```python
# Save cleaned dataset for downstream analysis
df.to_csv(
    "/drive/hospital-readmission-project/data/cleaned_hospital_data.csv",
    index=False
)
```
